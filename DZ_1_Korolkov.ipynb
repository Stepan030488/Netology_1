Задача 6
Напишите программу, которая сможет вычислять площади трех фигур (круг, треугольник и прямоугольник). Тип фигуры запрашиваем через пользовательский ввод, после чего делаем запрос характеристик фигуры:

```python
import math
while True:
    a=input('Назовите фигуру: круг, квадрат или треугольник: ')
    a=a.lower()
    if a=='круг':
        r=int(input('Введите радиус в см: '))
        print('Площадь круга равна: ', math.pi*(r**2))
        break
    elif a=='квадрат':
        r=int(input('Введите сторону в см: '))
        print('Площадь квадрата равна: ', r**2)
        break
    elif a=='треугольник':
        st1=int(input('Введите сторону 1 в см: '))
        st2=int(input('Введите сторону 2 в см: '))
        st3=int(input('Введите сторону 3 в см: '))
        p=(st1+st2+st3)/2
        print('Площадь треугольника равна: ', (p*(p-st1)*(p-st2)*(p-st3))**0.5)
        break
        
    else:
        print('Введите фигуру из списка')
```
